### TASK 1: Prediction using supervised ML.

##### Problem Statement: Predict the percentage of a student based on the no. of study hours (9.25hr/day).

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
score= pd.read_csv("student_scores - student_scores.csv")
```

```python
score.head()
```

```python
score.shape
```

```python
score.plot(x= 'Hours', y = 'Scores', style = 'o')
plt.title('Hours vs Percentage')
plt.xlabel('Hours Studied')
plt.ylabel('Marks percentage')
```

###### By observing the plot we can clearly see, strong linear relationship between the 2 variables and can imagine a straight trend line.

### DATA PREPARATION

```python
x = score.iloc[:,:-1].values
y = score.iloc[:,1].values
```

###### Now, we are splitting the data into training data and testing data in the ratio of 70:30 using train_test_split() method.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

#### TRAINING STAGE

```python
from sklearn.linear_model import LinearRegression
lrn = LinearRegression()
lrn.fit(x_train, y_train)
```

```python
line = lrn.coef_*x+lrn.intercept_

plt.scatter(x_train, y_train, color = 'red')
plt.plot(x, line)
plt.show()
```

```python
line = lrn.coef_*x+lrn.intercept_

plt.scatter(x, y, color = 'black')
plt.plot(x, line)
plt.show()
```

###### Making data predictions

```python
pr = lrn.predict(x_test)
```

```python
pr
```

#### Comparing the Actual vs Predicted data

```python
df = pd.DataFrame({'Actual' : y_test, 'Predicted': y_test})
df
```

#### INPUT THE PREDICTION

```python
hr = np.array(9.25)
hr = hr.reshape(-1,1)
pr = lrn.predict(hr)
```

#### CONCLUSION

```python
print("If the student studies for 9.25 hours, he/she is expected to score {} marks.".format(pr))
```
